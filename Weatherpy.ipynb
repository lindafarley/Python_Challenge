```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import requests as req
from citipy import citipy
import pickle
```

```python
cities = []

#Create list of 500 unique cities using citipy
while len(cities) < 500:
    lat = np.random.uniform(-90, 90)
    lon = np.random.uniform(-180, 180)
    city = citipy.nearest_city(lat,lon).city_name
    if city not in cities:
        cities.append(city)
```

```python
#Confirm number of unique cities
len(set((cities)))
```

```python
# Open Weather API key
apikey = "b688d85656296814d2a8ea78b7b7b00f"

#Create lists to store responses for temp, humidity, cloudiness, and windspeed
temperature = []
humidity = []
cloudiness = []
windspeed = []
city_name = []

#Loop through cities list, retrieve responses, and store in lists
for city in cities:
    url = "http://api.openweathermap.org/data/2.5/weather?appid=%s&q=%s" % (apikey, city)
    response = req.get(url).json()
    #If data not available, skip that city
    try:
        temperature.append(response["main"]["temp"])
        humidity.append(response["main"]["humidity"])
        cloudiness.append(response["clouds"]["all"])
        windspeed.append(response["wind"]["speed"])
        city_name.append(city)
        print("City Name: " + city + " City Number: " + str(response["id"]) + url)
    except:print("Skip " + str(city) + " due to missing field")
    
```

```python
#Data was not available for all cities
print(len(temperature))
print(len(humidity))
print(len(cloudiness))
print(len(windspeed))
print(len(city_name))
```

```python
#Create dataframe with city names, temps, humidities, cloudiness, and windspeed
df = pd.DataFrame({"City Name": city_name, 
              "Temperature": temperature, 
              "Humidity": humidity, 
              "Cloudiness": cloudiness, 
              "Windspeed": windspeed})

len(df)
```
